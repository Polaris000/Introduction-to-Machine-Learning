```python
from keras.models import Sequential
from keras.layers import Dense
import sys
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = '/home/icameisaticoded/Documents/Repos/Datasets'

sys.path.append(path)
import pandas as pd
import gradient_descent as gd

data = pd.read_csv(path + '/' + 'train.csv')
#print(data.shape)

data_test = pd.read_csv(path + '/' + 'test.csv')

X_train, y_train = gd.analyze_data(data)
X_train = X_train[:, 1]

X_test, y_test = gd.analyze_data(data_test)
X_test = X_test[:, 1]
print(X_train.shape)
```

```python
model = Sequential()
```

```python
layer = Dense(units=1, input_dim=1)
model.add(layer)
```

```python
model.compile(loss='mean_squared_error',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1000, batch_size=700)
```

```python
score = model.evaluate(X_test, y_test, batch_size=300)
```

```python
print(score)
```

```python
print(layer.input_shape, layer.output)
```

```python
y_pred = model.predict(X_test)
print(X_test.shape, y_pred.shape)
```
